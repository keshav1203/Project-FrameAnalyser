# **Color Spaces in OpenCV**

**In image processing, a color space defines how colors are represented in a digital image. OpenCV supports multiple color spaces, each suitable for different types of image analysis. When an image is loaded using OpenCV, it can be converted into different formats depending on the task**

# BGR (Blue–Green–Red)

BGR is the default color space used by OpenCV. Every pixel is represented using three values corresponding to blue, green, and red intensities. This format originated from early camera and scientific imaging standards, and OpenCV continues to use it for consistency. When images are loaded through OpenCV, they are stored in the BGR format.

# RGB (Red–Green–Blue)

RGB is the most widely used color model, commonly seen in computer graphics, display systems, and image viewers. Colors are represented through red, green, and blue channels, arranged in that order. Although RGB and BGR hold the same information, their channel order differs. Many visualization tools, including plotting libraries, expect images in RGB format, which is why conversion is often necessary before display.

# GRAYSCALE (Gray)

Grayscale is a single-channel image format where each pixel stores only intensity information instead of separate color channels. The intensity value represents brightness ranging from black to white. Because it requires only one channel, it reduces memory usage and computational cost, making it suitable for tasks like feature extraction, thresholding, edge detection, and other machine learning or computer vision operations. Grayscale images are especially useful in cases where color does not add significant value to the analysis.

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("kock1.jpg")


if image is None:
    print("Error: Image not found!")
    exit()


bgr = image


rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)


gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

resized_gray = cv2.resize(gray, (100, 100))


plt.figure(figsize=(10, 5))


plt.subplot(2, 2, 1)
plt.imshow(bgr)
plt.title("Original (BGR)")
plt.axis('off')


plt.subplot(2, 2, 2)
plt.imshow(rgb)
plt.title("RGB Image")
plt.axis('off')


plt.subplot(2, 2, 3)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale")
plt.axis('off')


plt.subplot(2, 2, 4)
plt.imshow(resized_gray, cmap='gray')
plt.title("Resized Grayscale (100×100)")
plt.axis('off')

plt.tight_layout()
plt.show()
```

# **YOLO Dataset**

YOLO, short for You Only Look Once, is one of the most widely used datasets and frameworks for real-time object detection. The YOLO dataset provides large-scale annotated images that train models to identify and locate objects in images or video streams. Unlike earlier detection approaches that required the image to be scanned multiple times, YOLO treats object detection as a single regression problem, making the detection process extremely fast and efficient.

The YOLO dataset typically contains thousands of images with labeled objects belonging to different classes, such as people, vehicles, animals, electronic items, and everyday objects. Each image in the dataset is annotated with bounding boxes and class labels that define what object is present and where it appears in the image. This structured annotation enables the model to learn not only how objects look, but also their spatial position within the image.

The YOLO dataset is designed to support real-time applications, and its training approach focuses on predicting object class and bounding box coordinates in a single forward pass of the neural network. This allows YOLO models to process multiple frames per second, making them highly suitable for real-time tasks such as surveillance systems, robotics, autonomous vehicles, gesture-based systems, and camera-based object tracking.

The dataset is commonly used with versions of the YOLO model such as YOLOv3, YOLOv5, YOLOv7, and YOLOv8, with each new version introducing improvements in accuracy, speed, feature extraction, and training capability. The YOLO dataset is also compatible with modern deep learning frameworks and libraries, making it easy for developers to customize, expand, or retrain for new object classes.

In summary, the YOLO dataset is a structured, large-scale image annotation set used to train YOLO models for real-time object detection. It provides the necessary labeled data for a model to learn object shapes, features, and spatial positioning, enabling reliable and high-speed detection in real-world applications.

```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt


model = YOLO("yolov8n.pt") 


image = cv2.imread("football.jpg")

if image is None:
    print("Image not found!")
    exit()


results = model.predict(image)


for result in results:
    boxes = result.boxes

    for box in boxes:
     
        x1, y1, x2, y2 = box.xyxy[0]

        x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)

        cls = int(box.cls[0])
        label = result.names[cls]
        conf = float(box.conf[0])

        cv2.rectangle(image, (x1, y1), (x2, y2), (0,255,0), 2)

        text = f"{label} {conf:.2f}"
        cv2.putText(image, text, (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    0.7, (0,255,0), 2)

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.title("Object Detection Using YOLO")
plt.axis('off')
plt.show()
```

```python
import cv2
import sys
import time
import numpy as np
from playsound import playsound
from ultralytics import YOLO

cv2.setNumThreads(1)

# Load YOLO pose model
model = YOLO("yolov8n-pose.pt")

# Alarm file
ALARM_SOUND = "beep.wav"

# Keypoints we want to label
KEYPOINT_MAPPING = {
    0: "Nose",
    1: "Left Eye",
    2: "Right Eye",
    3: "Left Ear",
    4: "Right Ear",
}

FONT = cv2.FONT_HERSHEY_SIMPLEX
FONT_SCALE = 0.5
FONT_COLOR = (0, 255, 255)
THICKNESS = 1
OFFSET_Y = -10

# ---------------------------------------------
# FUNCTION: Draw Keypoint Labels
# ---------------------------------------------
def draw_keypoint_labels(frame, keypoints, conf):
    for kp_index, kp_name in KEYPOINT_MAPPING.items():
        if conf[kp_index] < 0.50:  # confidence filtering
            continue

        x, y = map(int, keypoints[kp_index])

        cv2.putText(
            frame,
            kp_name,
            (x, y + OFFSET_Y),
            FONT,
            FONT_SCALE,
            FONT_COLOR,
            THICKNESS,
            cv2.LINE_AA
        )
        cv2.circle(frame, (x, y), 4, (0, 255, 255), -1)

# ---------------------------------------------
# FUNCTION: Extract Head Region
# ---------------------------------------------
def draw_head_region(frame, person_keypoints, conf):
    # Keep only head keypoints
    head_points = []

    for kp_index in KEYPOINT_MAPPING.keys():
        if conf[kp_index] >= 0.50:  
            head_points.append(person_keypoints[kp_index])

    if len(head_points) < 1:
        return False  # No head detected → alarm

    head_points = np.array(head_points)
    x_min, y_min = np.min(head_points, axis=0).astype(int)
    x_max, y_max = np.max(head_points, axis=0).astype(int)

    # Draw head box
    cv2.rectangle(
        frame,
        (x_min - 20, y_min - 20),
        (x_max + 20, y_max + 20),
        (255, 0, 0),
        2
    )
    cv2.putText(frame, "Head Region", (x_min, y_min - 25),
                FONT, 0.6, (255, 0, 0), 2)

    return True

# ---------------------------------------------
# MAIN
# ---------------------------------------------
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("❌ ERROR: Cannot access webcam.")
    sys.exit()

prev_time = 0
alarm_active = False

print("Press Q to quit...")

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame, stream=True)
    annotated_frame = frame.copy()

    head_detected_any = False

    for r in results:
        # YOLO bbox drawing
        if r.boxes:
            boxes = r.boxes.xyxy.cpu().numpy()
            for (x1, y1, x2, y2) in boxes:
                cv2.rectangle(annotated_frame, (int(x1), int(y1)),
                              (int(x2), int(y2)), (0, 255, 0), 2)
                cv2.putText(annotated_frame, "Person", (int(x1), int(y1) - 5),
                            FONT, 0.6, (0, 255, 0), 2)

        # Keypoints
        if r.keypoints and r.keypoints.data.numel() > 0:
            keypoints_xy = r.keypoints.xy.cpu().numpy()
            keypoints_conf = r.keypoints.conf.cpu().numpy()

            for i, person_keypoints in enumerate(keypoints_xy):
                conf = keypoints_conf[i]

                draw_keypoint_labels(annotated_frame, person_keypoints, conf)

                # Head region detection
                head_present = draw_head_region(
                    annotated_frame, person_keypoints, conf
                )

                if head_present:
                    head_detected_any = True

    # -----------------------------------------
    # Sound Alarm if head disappears
    # -----------------------------------------
    if not head_detected_any:
        if not alarm_active:
            alarm_active = True
            playsound(ALARM_SOUND, block=False)
            cv2.putText(annotated_frame, "⚠ HEAD NOT DETECTED!", (10, 450),
                        FONT, 0.8, (0, 0, 255), 3)
    else:
        alarm_active = False

    # -----------------------------------------
    # FPS Counter
    # -----------------------------------------
    curr_time = time.time()
    fps = 1 / (curr_time - prev_time) if prev_time != 0 else 0
    prev_time = curr_time

    cv2.putText(annotated_frame, f"FPS: {int(fps)}", (10, 30),
                FONT, 0.7, (0, 255, 0), 2)

    cv2.putText(annotated_frame, "Press 'Q' to Quit", (10, 470),
                FONT, 0.6, (255, 255, 255), 2)

    cv2.imshow("YOLOv8 Head Tracking + Alarm", annotated_frame)

    key = cv2.waitKey(1) & 0xFF
    if key == ord("q") or key == 27:
        break

cap.release()
cv2.destroyAllWindows()
sys.exit()
```
